```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

def load_instructions(save_path: str | Path) -> list[str]:
    with open(save_path, 'r') as f:
        instructions = f.readlines()
    return [s.strip() for s in instructions]
```

```python
for ind in range(20, 40):
    photo_dir = Path('../sampled', f'{ind}')
    sketch_dir = Path('../sampled', f'{ind}')
    removed_dir = Path('../inpainted', f'{ind}')
    photo_lst = list(photo_dir.glob('**/*_photo.jpg'))
    sketch_lst = list(sketch_dir.glob('**/*_sketch.png'))
    removed_lst = list(removed_dir.glob('**/*_remove.jpg'))
    if len(photo_lst) == 0 or len(sketch_lst) == 0 or len(removed_lst) == 0:
        continue
    photo_path = str(photo_lst[0])
    sketch_path = str(sketch_lst[0])
    removed_path = str(removed_lst[0])
    photo = Image.open(photo_path)
    sketch = Image.open(sketch_path)
    removed = Image.open(removed_path)
    fig, axes = plt.subplots(1, 3)
    axes[0].imshow(photo)
    axes[0].axis('off')
    axes[0].set_title('Photo')
    axes[1].imshow(sketch)
    axes[1].axis('off')
    axes[1].set_title('Sketch')
    axes[2].imshow(removed)
    axes[2].axis('off')
    axes[2].set_title('Removed')
    plt.show()
    instructions = load_instructions(f'../instructions/{ind}.txt')
    for instruction in instructions:
        print(instruction)
```
